```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

Reading the dataset


```python
game = pd.read_csv('/content/game.csv')
print(game.shape)
game.head(1730)
```

Checking for missing data in dataset


```python
game.isna().sum()
```

Category of games 

```python
base_color = sb.color_palette()[0]
gen_order = game['category'].value_counts().index

fig, ax = plt.subplots(figsize = (10,6))

sb.countplot(data = game, x= 'category', color = base_color, order = gen_order, 
             ax=ax)

plt.ylabel('number of downloads', fontsize = 14)
plt.xlabel('category of the game', fontsize = 14)
plt.xticks(rotation = 90);

```

```python
sorted_counts = game['category'].value_counts()

fig, ax = plt.subplots(figsize = (10,6))

pie = plt.pie(sorted_counts, labels = sorted_counts.index, startangle=90,
            counterclock = False)
plt.axis('square')

plt.title('most games in category', fontsize = 15);
```

```python
fig, ax = plt.subplots(figsize = (10,6))

plt.hist(data = game, x = 'category')

plt.ylabel('number of games ', fontsize = 12)
plt.xlabel('category of game', fontsize =12)
plt.xticks(rotation = 90);
```

Average rating


```python
base_color = sb.color_palette()[0]
gen_order = game['average rating'].value_counts().index

fig, ax = plt.subplots(figsize = (10,6))

sb.countplot(data = game, x= 'average rating', color = base_color, order = gen_order, 
             ax=ax)

plt.ylabel('number of apps', fontsize = 14)
plt.xlabel('average rating of the games', fontsize = 14)
plt.xticks(rotation = 90);
ax.set_ylim([0,50])
ax.set_xlim([0,10]);
```

```python
fig, ax = plt.subplots(figsize = (10,6))
plt.hist( data = game, x = 'average rating')

plt.ylabel('number of apps', fontsize = 12)
plt.xlabel('average rating of the games', fontsize = 12)
plt.xticks(rotation = 90);
```

```python
fig, ax = plt.subplots(figsize = (10,6))

sb.distplot(game['average rating'])
plt.ylabel('nunmber of apps', fontsize = 12)
plt.xlabel('average rating of the games', fontsize = 12);
```

60 days growth

```python
fig, ax = plt.subplots(figsize=  (10,6))

plt.hist(data = game, x = 'growth (60 days)')

plt.ylabel('number of apps', fontsize = 12)
plt.xlabel('growth of app in 60 days', fontsize = 12);

plt.xlim([1,500]);
plt.ylim([1,2000]);
```
